```python
import requests
import pandas as pd
import re
from datetime import datetime
pd.options.display.max_columns = 50

option_url = "https://www.mcxindia.com/backpage.aspx/GetOptionChain"
option_headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36",
    "Referer": "https://www.mcxindia.com/market-data/option-chain",
    "Accept-Language": "en-US,en;q=0.5",
    "Accept-Encoding": "gzip, deflate, br",
    "Connection": "keep-alive"
}
payload = {'Commodity':'CRUDEOIL','Expiry':'17JUL2023'} 
#payload = {'Commodity':'GOLD','Expiry':'26JUL2023'}

ocJson = requests.post(url = option_url, headers=option_headers,json=payload).json()
ocJson
```

```python
ocdf = pd.DataFrame(ocJson['d']['Data'])
ocdf
```

```python
ocdf = ocdf[(ocdf.CE_OpenInterest != 0) | (ocdf.PE_OpenInterest !=0 )]
ocdf
```

```python
ocdf.drop(['ExtensionData','ExpiryDate','Symbol' ,'CE_LTT','PE_LTT'], axis=1,inplace= True)
ocdf
```

```python
re.findall(r'\d+', "/Date(689088919)/dndkk8995")
```

```python
def parsedate(x):
    ltt = int(re.findall(r'\d+', x.LTT)[0])/1000
    return datetime.fromtimestamp(ltt)
```

```python
ocdf['LTT'] = ocdf.apply(parsedate , axis=1)
```
